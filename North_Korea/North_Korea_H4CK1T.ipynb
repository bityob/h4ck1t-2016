```python
open_2_close = { '(' : ')', '[' : ']', '{' : '}' } 
close_2_open = { ')' : '(', ']' : '[', '}' : '{' } 
```

```python
c = Counter(task)
c
```

```python
def solve_challange(task):   
    temp_task = task
    
    not_done = True
    
    while not_done:
        for i, t in enumerate(list(temp_task)):
            if (i >= (len(temp_task) - 1)):
                not_done = False

            if t not in close_2_open:
                continue

            c = Counter(temp_task[:i+1])

            if c[t] > c[close_2_open[t]]:
                temp_task = temp_task[:i] + close_2_open[t] + temp_task[i:]
                break

    not_done = True

    while not_done:
        for i, t in reversed(list(enumerate(temp_task))):
            if (i <= 0):
                not_done = False

            if t not in open_2_close:
                continue

            c = Counter(temp_task[i:])

            if c[t] > c[open_2_close[t]]:
                temp_task = temp_task[:i+1] + open_2_close[t] + temp_task[i+1:]
                break

    return temp_task
```

```python
solve = solve_challange(task)

Counter(solve)
```
